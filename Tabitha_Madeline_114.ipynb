```python
prices = soup.find_all('span', class_='price')
```

```python
data_reviewers = []
for each_reviewer in reviewers:
  data_reviewers.append(each_reviewer.text)
```

### Multiple Page

```python
data_products = []
data_prices = []
data_taxs = []
data_ratings = []
data_reviewers = []

for index in range(1,7):
  page = requests.get(f'https://shop.akasha.co.id/kategori-produk/perawatan-rambut/page/{index}/')
  soup = BeautifulSoup(page.content, 'html.parser')

  products = soup.find_all('h3', class_='name full_name')

  for each_product in products:
    data_products.append(each_product.text)

  prices = soup.find_all('span', class_='price')

  for each_price in prices:
    data_prices.append(each_price.text)

  ratings = soup.find_all('div', class_='star-rating')

  for each_rating in ratings:
    data_ratings.append(each_rating.text)

  taxs = soup.find_all('small', class_='woocommerce-price-suffix')

  for each_tax in taxs:
    data_taxs.append(each_tax.text)

  reviewers = soup.find_all('div', class_='count')

  for each_reviewer in reviewers:
    data_reviewers.append(each_reviewer.text)
```

```python
prices_fix = [prices[:9] for prices in data_prices]
```

```python
prices_fix
```

```python
soup
```

```python
data = {
    'Nama Products' : data_products,
    'Harga' : prices_fix,
    'Keterangan Pajak' : data_taxs,
    'Rating' : data_ratings,
    'Jumlah Reviewer' : data_reviewers
}
```

```python
df = pd.DataFrame.from_dict(data, orient='index')
df = df.transpose()
```

```python
print(len(data_products), len(data_prices), len(data_taxs), len(data_ratings), len(data_reviewers))
```

```python
pd.DataFrame(data).to_csv('akasha.csv')
```
